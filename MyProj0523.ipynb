```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import jieba
fp = open("alldata.txt","r",encoding="utf-8")
alldata = fp.read()

#載入停用詞，放到stopwords串列中
with open("stopwords.txt", "r", encoding="utf-8") as fp:
    stopwords = fp.readlines()
    stopwords = [w.strip() for w in stopwords]
print(stopwords)
tt = jieba.cut(alldata)
terms = list()
for t in tt:
    if t not in stopwords:
        terms.append(t)
    
alldata = ",".join(terms)


mask = np.array(Image.open('cloud.jpg'))
wordcloud = WordCloud(background_color="white", width=1000, height=860, margin=2, font_path="SimHei.ttf", mask=mask).generate(alldata)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
